```python
recent_grads.plot(kind = 'scatter' , x = 'Sample_size',
                 y = 'Median', )
```

Its rather hard to tell if, as the sample_size increases, the median income also increases. Perhaps it is the distribution of the graph. But most of the ss are under 1000 responses, and it goes up linearly from there. As we move laterally to the right, we don't see spikes in income. 

```python
recent_grads.plot(kind = 'scatter' , x = 'Sample_size',
                 y = 'Unemployment_rate')
```

```python
recent_grads.plot(kind = 'scatter' , x = 'Full_time',
                 y = 'Median')
```

As with the first question, it is difficult to ascertain whether, as the number of full_time of increases, a pattern is created with the Median income. Most of the fulltime responses are clustered under 50000, and kind of go up with a slope of 1. 

```python
recent_grads.plot(kind = 'scatter' , x = 'ShareWomen',
                 y = 'Unemployment_rate')
```

```python
recent_grads.plot(kind = 'scatter' , x = 'Men',
                 y = 'Median')
```

```python
recent_grads.plot(kind = 'scatter' , x = 'Women',
                 y = 'Median')
```

```python
recent_grads.plot(kind = 'scatter' , x = 'ShareWomen',
                 y = 'Median')
```

This fairly crude graph suggests that, as the share of women increases within the major, the median income actually decreases. Essentially, there is a negative linear relationship between income and the proportion of women in a particular major.
